```python
path = '/content/drive/My Drive/Colab Notebooks/EPAM DS foundations course/DS_module3/Australian cities metadata.xlsx'
cities_meta = pd.read_excel(path)
cities_meta.set_index('City Name', inplace=True)
cities_meta.head()
```

```python
# drop the second Climate zone column
cities_meta = cities_meta.drop(columns=['Climatic Zone 2'])
```

```python
# rename column name Climatic zone 1 to zone
cities_meta = cities_meta.rename(columns={'Climate Zone 1': 'Climatic zone'})
```

```python
cities_meta[['Latitude', 'Longitude']] = cities_meta['Coordinates'].str.split(',', expand=True)
```

```python
cities_meta.head()
```

```python
# remove characters except numbers from columns
cities_meta['Latitude'] = cities_meta['Latitude'].str.replace(r'[^\d.]', '', regex=True)
cities_meta['Longitude'] = cities_meta['Longitude'].str.replace(r'[^\d.]', '', regex=True)
```

```python
# Convert latitude and longitude to numeric types, handling potential errors
cities_meta['Latitude'] = pd.to_numeric(cities_meta['Latitude'], errors='coerce')
cities_meta['Longitude'] = pd.to_numeric(cities_meta['Longitude'], errors='coerce')

cities_meta.head()
```

```python
cities_meta.info()
```

```python
cities_meta = cities_meta.drop(columns=['Coordinates'])
```
